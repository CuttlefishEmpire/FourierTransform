```python
from scipy.io import wavfile
import numpy as np

# UPTOWN FUNK CELL

AudioName = "Uptown Funk.wav" # Audio file name
fs, Audiodata = wavfile.read(AudioName) 

Audiodata = np.mean(Audiodata, axis = 1) # Converts from stereo

# Plot the audio signal in time
import matplotlib.pyplot as plt
plt.plot(Audiodata)
plt.title('Audio signal in time - Uptown Funk',size=16)

# Power spectrum
from scipy.fftpack import fft
n = len(Audiodata) 
AudioFreq = fft(Audiodata)
AudioFreq = AudioFreq[0:int(np.ceil((n+1)/2.0))]
MagFreq = np.abs(AudioFreq) # Magnitude
MagFreq = MagFreq / float(n)
MagFreq = MagFreq**2
if n % 2 > 0: # ffte odd 
    MagFreq[1:len(MagFreq)] = MagFreq[1:len(MagFreq)] * 2
else:# fft even
    MagFreq[1:len(MagFreq) -1] = MagFreq[1:len(MagFreq) - 1] * 2 

# Remove negative volumes, scale volumes appropriately
noNegs = 10*np.log10(MagFreq[:len(MagFreq)//4])
noNegs2 = []
for neg in noNegs:
    if neg > 0:
        noNegs2.append(neg)
    else:
        noNegs2.append(0)

plt.figure()
freqAxis = np.arange(0,int(np.ceil((n+1)/2.0)), 1.0) * (fs / n);
parsedFreqAxis = freqAxis[:len(freqAxis)//4]
boundaryListUnparsed = [0]
boundaryList = []
maxList = []
maxElement = 0
for f, v in zip(parsedFreqAxis, noNegs2):
    if v > 25:
        maxList.append(v)
    else:
        for m in maxList:
            if m > maxElement:
                maxElement = m
        if maxElement != 0:
                boundaryListUnparsed.append(int(f))
        maxList = []
        maxElement = 0
for e in range(1, len(boundaryListUnparsed)):
    if boundaryListUnparsed[e] - boundaryListUnparsed[e-1] > 10 or boundaryListUnparsed[e-1] - boundaryListUnparsed[e] > 15:
        boundaryList.append(boundaryListUnparsed[e])
plt.plot(parsedFreqAxis,noNegs2)
plt.xlabel('Frequency (Hz)'); plt.ylabel('Power spectrum (dB)')

plt.show()
```
